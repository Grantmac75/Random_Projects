## Regular Expression & Read
Note that f.read() reads the whole file into memory so it's not as memory efficient.

```python
import re
from ast import literal_eval

pattern = re.compile(r"\('\D\d.*'\)")

with open('/Users/kevin/Dropbox/Github/DataCamp/Edges.rtf', 'r', encoding='utf-8') as f:
    
    contents = f.read()
    matches = pattern.findall(contents.strip('"'))
    
    some_list = []
    for match in matches:
        #print(match)
        some_list.append(literal_eval(match))

some_list
```
